# **Data Visualization Techniques — Iris Dataset**

Objective:
Explore the Iris dataset using different visualization techniques.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.info()
df.describe()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.show()
```

```python
df.hist(figsize=(10,6))
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.violinplot(x='species', y='petal_length', data=df)
plt.show()
```

```python
sns.barplot(x='species', y='sepal_length', data=df)
plt.show()
```

**Conclusion**:Various visualization techniques such as scatter plots, histograms, heatmaps, violin plots, and bar charts were used to analyze the Iris dataset. The species are clearly separable based on petal dimensions.
